```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("/content/TRAQID.csv")
df.head(5)
```

```python
df.Day_or_Night.value_counts()
```

```python
df.isnull().sum() #no missing values
```

change labels into dummy values


```python
df.replace('Satisfa','Satisfactory',inplace=True)
df.replace('Satisfactoryy','Satisfactory',inplace=True)
df.replace('Very Poor','Very_Poor',inplace=True)
```

```python
df1=pd.get_dummies(data=df,columns=['Day_or_Night','Season','aqi_cat'])
df1.Day_or_Night_Day=df1.Day_or_Night_Day.astype(np.float64)
df1.Day_or_Night_Night=df1.Day_or_Night_Night.astype(np.float64)
df1.Season_Monsoon=df1.Season_Monsoon.astype(np.float64)
df1.Season_Winter=df1.Season_Winter.astype(np.float64)
df1.aqi_cat_Good=df1.aqi_cat_Good.astype(np.float64)
df1.aqi_cat_Moderate=df1.aqi_cat_Moderate.astype(np.float64)
df1.aqi_cat_Poor=df1.aqi_cat_Poor.astype(np.float64)
df1.aqi_cat_Satisfactory=df1.aqi_cat_Satisfactory.astype(np.float64)
df1.aqi_cat_Severe=df1.aqi_cat_Severe.astype(np.float64)
df1.aqi_cat_Very_Poor=df1.aqi_cat_Very_Poor.astype(np.float64)
df1.head(5)
```

```python
cols_to_scale=['Temperature','Humidity','PM2.5','PM10','aqi']
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df1[cols_to_scale]=scaler.fit_transform(df1[cols_to_scale])
df1.sample(5)
```

```python
X=df1.drop(['created_at','Image','aqi'],axis=1)
y=df1['aqi']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.8,random_state=5)
```

```python
model=keras.Sequential([
    keras.layers.Dense(15,input_shape=(15,),activation='relu'),
    keras.layers.Dense(100,activation='relu'),
    keras.layers.Dense(1,activation='linear') # Changed activation to linear for regression
])
model.compile(optimizer='adam',loss='mean_squared_error',metrics=['mean_absolute_error']) # Changed loss and metrics for regression
model.fit(X_train,y_train,epochs=100)
```

```python
model.evaluate(X_train,y_train)
```

```python
model.evaluate(X_test,y_test)
```
